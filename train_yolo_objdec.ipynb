```python
from ultralytics import YOLO
import yaml
import os

# Load the YOLOv8 Nano segmentation model
model = YOLO('yolov8n-seg.pt')  # 'yolov8n-seg.pt' is the YOLOv8 Nano model for segmentation

# Define paths to your datasets
base_path = '/home/tarekz/Desktop/susu_summer24/factory_dataset/yolo_label/dataset_v1/'
train_images = os.path.join(base_path, 'images/train')
val_images = os.path.join(base_path, 'images/val')
test_images = os.path.join(base_path, 'images/test')

train_labels = os.path.join(base_path, 'labels/train')
val_labels = os.path.join(base_path, 'labels/val')
test_labels = os.path.join(base_path, 'labels/test')

# Ensure the paths exist
assert os.path.exists(train_images), f"Training images path not found: {train_images}"
assert os.path.exists(val_images), f"Validation images path not found: {val_images}"
assert os.path.exists(test_images), f"Test images path not found: {test_images}"
assert os.path.exists(train_labels), f"Training labels path not found: {train_labels}"
assert os.path.exists(val_labels), f"Validation labels path not found: {val_labels}"
assert os.path.exists(test_labels), f"Test labels path not found: {test_labels}"

# Define number of classes and class names
nc = 2  # number of classes
names = ["folding_machine", "worker"]  # list of class names

# Create the dataset dictionary for segmentation
dataset = {
    'path': base_path,
    'train': 'images/train',
    'val': 'images/val',
    'test': 'images/test' if os.path.exists(test_images) else '',  # Add test path only if it exists
    'names': names,
    'nc': nc
}

# Save the dataset dictionary to a temporary YAML file
with open('temp_data.yaml', 'w') as outfile:
    yaml.dump(dataset, outfile, default_flow_style=False)

# Specify the directory to save the trained model
save_dir = '/home/tarekz/Desktop/susu_summer24/trained_models/semantic_segmentation/'

# Train the model for segmentation
try:
    model.train(data='temp_data.yaml', epochs=100, imgsz=640, device='cuda', task='segment', save_dir=save_dir)

    # Evaluate the model on the test dataset
    if os.path.exists(test_images) and os.path.exists(test_labels):
        results = model.val(data='temp_data.yaml', split='test')
        print("Test evaluation results:", results)
finally:
    # Optionally, remove the temporary YAML file after training
    if os.path.exists('temp_data.yaml'):
        os.remove('temp_data.yaml')

# Print the paths to the saved models
best_model_path = os.path.join(save_dir, 'weights/best.pt')
last_model_path = os.path.join(save_dir, 'weights/last.pt')
print(f"Best model saved at: {best_model_path}")
print(f"Last model saved at: {last_model_path}")
```
